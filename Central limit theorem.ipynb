```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

## Central limit theorem excercise
Show how sampling from a normal or non-normal distribution becomes aproximately normal

```python
adist = stats.norm(loc=50,scale=5)
line = np.linspace(-4,4,200)
sns.kdeplot(adist.ppf(line))
```

```python
adist = stats.norm(loc=50,scale=5)
rvariates = adist.rvs(1000)
rv_np = np.array(rvariates)
```

```python
as_matrix = np.reshape(rv_np,(100,10))
r_means = np.mean(as_matrix,1)
sns.kdeplot(r_means)
```

```python
bdist = stats.poisson(mu=10)
line2 = (np.linspace(0,200,200))
line2int = line2.astype(int)
sns.kdeplot(bdist.pmf(line2int))
```

```python
rvariates2 = bdist.rvs(1000)
rv_np2 = np.array(rvariates2)
as_matrix2 = np.reshape(rv_np2,(100,10))
r_means2 = np.mean(as_matrix2,1)
sns.kdeplot(r_means2)
```
